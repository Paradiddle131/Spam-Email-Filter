### Spam Email Filter By Using Random Forest Classifier Algorithm From Sklearn Library

Got the email dataset from:

https://www.kaggle.com/balakishan77/spam-or-ham-email-classification/data

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df = pd.read_csv("emails.csv")
```

```python
df.head()
```

1: Spam

0: Ham (Not spam)

```python
with pd.option_context('display.max_rows', 5728):
    print (df)
```

```python
X, y = make_classification(n_samples=2736, n_features=2, n_informative=2, n_redundant=0, random_state=5)
```

```python
X
```

```python
y
```

```python
# split the data into 70% training, 30% test portions for future validations
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
classifier = RandomForestClassifier(n_estimators=20, random_state=5)
classifier
```

```python
classifier.fit(X, y)
```

```python
print(classifier.feature_importances_)
```

```python
print(classifier.predict([[0, 0]]))
```

```python
y_prediction = classifier.predict(X_test)
y_prediction
```

```python
# Accuracy
print(metrics.accuracy_score(y_test, y_prediction))
```
